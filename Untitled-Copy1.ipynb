## 데이터프레임 연산

```python
import IPython
from IPython.display import display_html
```

```python
def display_by_side(*args):
    # 두 개 데이터프레임을 옆으로 나란히 표시하는 함수
    html_str = ''
    for df in args:
        html_str += df.to_html()
    display_html(html_str.replace('table','table style="display:inline"'), raw=True)
```

```python
# +=, -=: 증감연산자, a += b <-> a = a + b
x = 5
y = 5
x += 5
y = y + 5
print(x, y)
```

```python
df1 = pd.DataFrame({'KOR':[1, 3], 'USA':[2, 4]}, index=['2018', '2019'])
df2 = df1 + 10
df2.columns = ['USA', 'CHN']
display_by_side(df1, df2, df1 + df2) # 라벨이 같지 않으면, NaN 처리
```

### f-string을 이용한 x, y축 값 표시 형식 지정

```python
from matplotlib.ticker import FuncFormatter as FF
from matplotlib.dates import DateFormatter as DF
from matplotlib.dates import DayLocator as DL
```

```python
fig, ax = plt.subplots(1, 2, figsize=(8, 3))
df.plot(ax=ax[0])
df.iloc[2:].plot(ax=ax[1])
ax[1].legend(loc=4, fontsize=18)
dp2 = lambda x, p: f'{x:,.2f}'
ax[1].yaxis.set_major_formatter(FF(dp2)) # FF는 인자가 함수
```

```python
fig, ax = plt.subplots(1, 2, figsize=(8, 3))
ax[0].plot(df.index, df)
ax[1].plot(df.index[2:], df.iloc[2:])
#ax[1].legend(loc=4, fontsize=18)
dp2 = lambda x, p: f'{x:,.1f}'
ax[1].yaxis.set_major_formatter(FF(dp2)) # FF는 인자가 함수
ax[1].xaxis.set_major_formatter(DF('%m-%d'))
ax[1].xaxis.set_major_locator(DL(interval=1))
```

```python
def xaxis_date_format(ax, fmt='%m-%d'):
    ticklabels = [pd.to_datetime(t.get_text()).strftime(fmt) for t in ax.get_xticklabels()]
    ax.set_xticklabels(ticklabels, rotation=0)
```

```python
fig, axs= plt.subplots(1, 1, figsize=(8, 4))

df.plot.bar(ax=axs)

xaxis_date_format(axs, '%m-%d')

plt.legend()
plt.title('BAR chart')
```
